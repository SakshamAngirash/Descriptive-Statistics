```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Iris.csv")
```

```python
data.rename(columns={"SepalLengthCm":"SL","SepalWidthCm":"SW","PetalLengthCm":"PL","PetalWidthCm":"PW"},inplace=True)
```

```python
data.head()
```

```python
iris_s=data[data["Species"]=="Iris-setosa"]
iris_v=data[data["Species"]=="Iris-versicolor"]
iris_vi=data[data["Species"]=="Iris-virginica"]
```

```python
sns.distplot(iris_s["PL"])
```

```python
sns.FacetGrid(data,hue="Species",height=6).map(sns.distplot,"PL").add_legend()
```

```python
counts,bin_edges=np.histogram(iris_s["PL"],bins=10)
```

```python
pdf=counts/sum(counts)
pdf
```

```python
cdf=np.cumsum(pdf)
cdf
```

```python
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
counts,bin_edges=np.histogram(iris_s["PL"],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts,bin_edges=np.histogram(iris_v["PL"],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts,bin_edges=np.histogram(iris_vi["PL"],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

```
